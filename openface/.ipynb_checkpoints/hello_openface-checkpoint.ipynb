```python
import cv2
import numpy as np
np.set_printoptions(precision=2)
import openface
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
fileDir = "/Users/zzh/study/gitProjects/openface"
modelDir = os.path.join(fileDir, 'models')
dlibModelDir = os.path.join(modelDir, 'dlib')
openfaceModelDir = os.path.join(modelDir, 'openface')
```

```python
fileDir
```

```python
imgPath1=os.path.join(fileDir,"images/examples/lennon-1.jpg")
imgPath2=os.path.join(fileDir,"images/examples/lennon-2.jpg")
dlibFacePredictor=os.path.join(dlibModelDir, "shape_predictor_68_face_landmarks.dat")
networkModel=os.path.join(openfaceModelDir, 'nn4.small2.v1.t7')
imgDim=96
```

```python
align = openface.AlignDlib(dlibFacePredictor)
```

```python
net = openface.TorchNeuralNet(networkModel, imgDim)
```

```python
bgrImg1=cv2.imread(imgPath1)
plt.imshow(bgrImg1)
```

```python
rgbImg1=cv2.cvtColor(bgrImg1, cv2.COLOR_BGR2RGB)
plt.imshow(rgbImg1)
```

```python
bb1 = align.getLargestFaceBoundingBox(rgbImg1)
bb1
```

```python
landmarks1=align.findLandmarks(rgbImg1,bb1)
```

```python
rgbImg1LandMarked=rgbImg1.copy()
```

```python
for landmark in landmarks1:
    cv2.circle(rgbImg1LandMarked,landmark,5,(55,255,155),5)
```

```python
plt.imshow(rgbImg1LandMarked)
```

```python
alignedFace1 = align.align(imgDim, rgbImg1, bb1,landmarkIndices=openface.AlignDlib.OUTER_EYES_AND_NOSE)
```

```python
plt.imshow(alignedFace1)
```

```python
rep1 = net.forward(alignedFace1)
```

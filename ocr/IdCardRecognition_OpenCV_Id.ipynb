```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
%matplotlib inline
```

```python
img=cv2.imread("img/IDCard001.png")
# img=cv2.imread("D:\\tmp\\IDCard002.jpg")
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
